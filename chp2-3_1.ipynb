```python
class Complex:
    """ complex number class """
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
        
    def __repr__(self):
        if self.imag < 0:
             return f'{self.real}{self.imag}i'
        return f'{self.real}+{self.imag}i'
        
    def modulus(self):
        """ return modulus """
        return (self.real**2 + self.imag**2)**0.5
    
    def arg(self):
        """ return argument, in rad """
        return atan2(self.imag, self.real)
    
    def pow(self, n):
        """ return power n """
        r = self.modulus()
        a = self.arg()
        return Complex(r**n * cos(n * a), r**n * sin(n * a))
    
    def conjugate(self):
        """ return conjugate """
        return Complex(self.real, -self.imag)
```

### 运算符重载

```python
class Complex:
    """ complex number class """
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
        
    def __repr__(self):
        if self.imag < 0:
             return f'({self.real}{self.imag}i)'
        return f'({self.real}+{self.imag}i)'
        
    def __abs__(self):
        """ return absolute """
        return self.modulus()
    
    def __neg__(self):
        """ return negative complex """
        return Complex(-self.real, -self.imag)
    
    def __add__(self, c):
        """ return self + c """
        return Complex(self.real + c.real, self.imag + c.imag)
    
    def __sub__(self, c):
        """ return self - c """
        return Complex(self.real - c.real, self.imag - c.imag)
    
    def __mul__(self, c):
        """ return self * c """ 
        return Complex(self.real * c.real - self.imag * c.imag,
                      self.real * c.imag + self.imag * c.real)
    
    def __truediv__(self, c):
        """ return self/c """
        t = c.real**2 + c.imag**2
        return Complex((self.real * c.real + self.imag * c.imag) / t,
                      (self.imag * c.real - self.real * c.imag) / t)
        
    def __pow__(self, n):
        """ return power n """
        return self.pow(n)
    
    def __eq__(self, c):
        """ if self == c """
        return self.real == c.real and self.imag == c.imag    
        
    def modulus(self):
        """ return modulus """
        return (self.real**2 + self.imag**2)**0.5
    
    def arg(self):
        """ return argument, in rad """
        return atan2(self.imag, self.real)
    
    def pow(self, n):
        """ return power n """
        r = self.modulus()
        a = self.arg()
        return Complex(r**n * cos(n * a), r**n * sin(n * a))
    
    def conjugate(self):
        """ return conjugate """
        return Complex(self.real, -self.imag)
```

### 类方法和静态方法

```python
class Dog:
    category = 'dog'
    def __init__(self, name, age):
        self.name = name
        self.age = age
       
    @classmethod
    def shout(cls):
        return 'Bark'
    
    @staticmethod
    def compare_age(pet1, pet2):
        if pet1.age < pet2.age:
            return f'{pet1.name} is younger than {pet2.name}.'
        elif pet1.age == pet2.age:
            return f'{pet1.name} is same old as {pet2.name}.'
        else:
            return f'{pet1.name} is older than {pet2.name}'
    
    def introduce(self):  # 注意这里的换行
        return (f'{Dog.shout()}, my name is {self.name}, '
                f'I am {self.age} years old, I am {self.category}.')
```

### 内部变量和setter


```python
class Dog:
    category = 'dog'
    def __init__(self, name, age):
        self.name = name
        self._age = age
       
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, age):
        if age < 0: print('age can not be negative.')
        else: self._age = age
        
    def shout(self):
        return 'Bark'
    
    def introduce(self):  # 注意这里的换行
        return (f'{self.shout()}, my name is {self.name}, '
                f'I am {self._age} years old, I am {self.category}.')
```

### 多态

设想这样的情形：森林中召开动物联欢会，有一个环节是各个动物都要做自我介绍。

```python
def introduce_section(animals):
    for m in animals:
        print(m.introduce())
# animals 是动物的对象列表
# 要求每个 animal 都有 introduce 方法
```

```python
# 设计一个 Animal 类，作为所有动物都基类，不同种类的动物从 Animal 类继承
class Animal:
    category = 'animal'
    def __init__(self, name, age):
        self.name = name
        self._age = age
       
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, age):
        if age < 0: print('age can not be negative.')
        else: self._age = age
        
    def shout(self):
        return '....'
    
    def introduce(self):  # 注意这里的换行
        return (f'{self.shout()}, my name is {self.name}, '
                f'I am {self._age} years old, I am {self.category}.')
```

```python
class Pig(Animal):
    category = 'pig'
    def __init__(self, name, age):
        super().__init__(name, age)
        
    def shout(self):
        return 'hennn'
```

```python
class Tiger(Animal):
    category = 'tiger'
    def __init__(self, name, age):
        super().__init__(name, age)
        
    def shout(self):
        return 'mooo'
```

```python
class Goat(Animal):
    category = 'goat'
    def __init__(self, name, age):
        super().__init__(name, age)
        
    def shout(self):
        return 'Maaah'
```

```python
animals = [Goat('Suzy', 3), Tiger('Freddy', 7), Pig('Peppa', 5)]
introduce_section(animals)
```

### duck-type, 鸭子类型


```python
class Duck:
    def __init__(self, name, age):
        self.name, self.age = name, age
    
    def introduce(self):
        return f'qurck, I am a duck, I am {self.age} years old, my name is {self.name}.'
```

```python
animals.append(Duck('Molly', 2))
introduce_section(animals)
```

Duck 没有从 Animal 继承，但是因为它也有  introduce 方法，而且 函数 introduce_section 并不检查/要求 参数 animals 中元素/对象的类型，只要调用对象的 introduce 方法不出错就可以。

鸭子类型的特性在 Python 中应用广泛。
